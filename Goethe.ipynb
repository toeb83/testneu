```python
import urllib.request
import re
```

```python
u2 = urllib.request.urlopen('http://www.digbib.org/Johann_Wolfgang_von_Goethe_1749/Faust_I?textonly=1')
mystr=""
regex = re.compile('[;|,|:|.|(|)|!|?|-]')
regex2 = re.compile('\'s')
for lines in u2.readlines():
    lines=str(lines, 'utf-8', errors='replace')
    lines=lines.replace("\n",' ')
    lines=regex.sub(' ', lines )
    lines=regex2.sub(' es', lines )
    lines = lines.strip()
    mystr+=lines
```

```python
print (mystr)
```

```python
mystr=mystr.lower()
```
